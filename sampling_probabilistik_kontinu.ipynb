Energi neutron yang dihasilkan dari reaksi fusi memenuhi distribusi probabilitas Watt.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def P(E):
  return (0.4865 * np.sinh(np.sqrt(2*E))*np.exp(-E))

E = np.arange(0, 10, 0.01)
p = P(E)

n = 100000
acc = []
for i in range(n):
  energy = np.random.default_rng().uniform(0,10,1)
  prob_func = np.random.default_rng().uniform(0,np.max(p),1)
  if prob_func <= P(energy):
    acc.append(energy[0])
  else:
    pass

plt.hist(acc, bins=100)
plt.show()
```
