$\Delta G = -\frac{4\pi RTr^3}{3V_m} . \ln{\frac{P}{P_0^K}}+4\pi \gamma r^2$

```python
# Rio Agustian Gilang Fernando
# Advanced Material Science and Nanotechnology
# University of Science and Technology of Hanoi

import numpy as np
import matplotlib.pyplot as plt

def gibbs_energy(r, ratio_p):
  r = r * 1e-9
  gamma = 0.072 #Jm^-2
  Vm = 18 * 1e-6 #m^3
  R = 8.314
  T = 293.15 #Kelvin
  expr_1 = - (4 * np.pi * R * T * r**3) / (3 * Vm)
  expr_2 = np.log(ratio_p)
  expr_3 = 4 * np.pi * gamma * r**2
  return (expr_1 * expr_2 + expr_3) / (4.05 * 1e-21)

x = np.linspace(0, 3, 500)
y = gibbs_energy(x, 2)
y1 = gibbs_energy(x, 3)
y2 = gibbs_energy(x, 4)

plt.plot(x, y, label="$P/P_0=2$")
plt.plot(x, y1, label="$P/P_0=3$")
plt.plot(x, y2, label="$P/P_0=4$")
plt.ylim(0, max(y) + 10)
plt.ylabel("$\Delta G (k_B T)$")
plt.xlabel("Radius (nm)")
plt.legend()
plt.grid()
plt.show()
```
