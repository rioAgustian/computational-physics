## Absorbansi

```python
import numpy as np
import matplotlib.pyplot as plt

path = r"C:\Users\Rio Agustian\Downloads\data.csv"
data = np.loadtxt(path, delimiter=";", dtype=str)
wvl = data[1:, 0].astype(np.float64)
ab1 = data[1:, 1].astype(np.float64)
ab2 = data[1:, 2].astype(np.float64)

plt.figure(figsize=(8,4))
plt.plot(wvl, ab1, lw=1, label="400°C")
plt.plot(wvl, ab2, lw=1, label="600°C")
plt.xlabel("Panjang gelombang (nm)")
plt.ylabel("Absorbansi")
plt.legend()
plt.show()
```

## Transmitansi

```python
tr1 = np.max(ab2) - ab1
tr2 = np.max(ab2) - ab2

plt.figure(figsize=(8,4))
plt.plot(wvl, tr1, lw=1, label="400°C")
plt.plot(wvl, tr2, lw=1, label="600°C")
plt.xlabel("Panjang gelombang (nm)")
plt.ylabel("Transmitansi")
plt.legend()
plt.show()
```

## Absorbansi 300-1200

```python
plt.figure(figsize=(8,4))
plt.plot(wvl[100:], ab1[100:], lw=1, label="400°C")
plt.plot(wvl[100:], ab2[100:], lw=1, label="600°C")
plt.xticks(ticks=np.arange(300, 1300, 100))
plt.xlabel("Panjang gelombang (nm)")
plt.ylabel("Absorbansi")
plt.legend()
plt.show()
```
