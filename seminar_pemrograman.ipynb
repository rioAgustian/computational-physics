## **Perhitungan Panjang**

$i = \frac{V}{R}e^{-t/RC}$

```python
# Kode ini digunakan untuk menghitung arus (I) yang mengalir pada rangkaian RC

# Library
import numpy as np

# Parameter
V = 10
R = 100
C = 1e-5
t = R*C

# Perhitungan
i = (V/R) * np.exp(-t/(R*C))

# Keluaran
i
```

## **Function**

```python
# Function
def arus (V, R, C, t):
  i = (V/R) * np.exp(-t/(R*C))
  return i

# Parameter
V = 10
R = 100
C = 1e-5
t = R*C

# i1 = arus(10, 100, 1e-5, t)
i2 = arus(V, R, C, t)
```

## **Loop**

```python
# while loop
dt = 1   # step
t  = 0     # start
t_max = 5  # stop

while t <= t_max :
  # operasi di sini
  print(t)
  t = t + dt
```

```python
# Parameter fisis
V = 10
R = 100
C = 1e-5

# Parameter loop
dt = R*C / 10
t  = 0.0
t_max = 4 * R * C

# Loop
while t <= t_max :

  # Perhitungan
  i = arus(V, R, C, t)

  # Keluaran
  print(i)

  # Update t
  t = t + dt
```

## **Menyimpan Data**

```python
# Parameter fisis
V = 10
R = 100
C = 1e-5

# Parameter loop
dt = R*C / 1
t  = 0.0
t_max = 6 * R * C

# Tempat penyimpanan
curr = []
time = []

# Loop
while t <= t_max :

  # Perhitungan
  i = arus(V, R, C, t)

  # Menyimpan data
  curr.append(i)
  time.append(t)

  # Update t
  t = t + dt

curr
```

## **Visualisasi Data**

```python
import matplotlib.pyplot as plt

plt.plot(time, curr)
plt.show()
```

## **Fine Tunning**


```python
plt.plot(time, curr,
         color="black",
         label="Current flow")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Current flow through capacitor as a function of time")
plt.legend()
plt.show()
```

```python
plt.scatter(time, curr,
            color="black",
            marker = "s",
            s = 100,
            label="Data points")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Current flow through capacitor as a function of time")
plt.grid()
plt.legend()
plt.show()
```

## **Dua Grafik Bersebelahan**

```python
plt.figure(figsize=(10, 4))

# Grafik 1
plt.subplot(121) # baris - kolom - gambar ke)
plt.plot(time, curr,
         color="black",
         label="Current flow")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.legend()

# Grafik 2
plt.subplot(122)
plt.scatter(time, curr,
            color="black",
            marker = "s",
            label="Data points")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## **Gabungkan!**

```python
# Parameter fisis
V = 10
R = 100
C = 1e-5

# Parameter loop
dt = R*C / 10
t  = 0.0
t_max = 6 * R * C

# Tempat penyimpanan
curr = []
time = []

# Loop
while t <= t_max :

  # Perhitungan
  i = arus(V, R, C, t)

  # Menyimpan data
  curr.append(i)
  time.append(t)

  # Update t
  t = t + dt

plt.figure(figsize=(10, 4))

plt.subplot(121) # baris - kolom - gambar ke)
plt.plot(time, curr,
         color="black",
         label="Current flow")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.legend()

plt.subplot(122)
plt.scatter(time, curr,
            color="black",
            marker = "s",
            label="Data points")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## **Versi pendek**


```python
# Parameter fisis
V = 10
R = 100
C = 1e-5

time = np.arange(0, 6*C*R, R*C/100)
curr = arus(V, R, C, time)

# plotting

plt.figure(figsize=(10, 4))

plt.subplot(121) # baris - kolom - gambar ke)
plt.plot(time, curr,
         color="black",
         label="Current flow")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.legend()

plt.subplot(122)
plt.scatter(time, curr,
            color="black",
            marker = "s",
            label="Data points")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
