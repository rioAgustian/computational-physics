```python
# Rio Agustian Gilang Fernando
# Advanced Material Science and Nanotechnology

import numpy as np
import matplotlib.pyplot as plt

def current(E, alpha):
  current_0 = 1e-6
  F = 96485
  R = 8.314
  T = 293
  eta = (E - 0)
  first_expr  = np.exp((1-alpha) * F/(R*T) * eta)
  second_expr = np.exp(-alpha * F/(R*T) * eta)
  return current_0 * (first_expr - second_expr)

x = np.linspace(-1, 1, 500)
y = current(x, 0.4)
y1 = current(x, 0.5)
y2 = current(x, 0.6)

plt.plot(x, y, label=r"$\alpha_1 = 0.4$")
plt.plot(x, y1, label=r"$\alpha_2 = 0.5$")
plt.plot(x, y2, label=r"$\alpha_3 = 0.6$")
plt.xlabel("E (volt)")
plt.ylabel("i (ampere)")
plt.title("Volmer-Butler Equation")
plt.ylim(-1, 1)
plt.vlines(0, min(y), max(y), colors="black", linestyles="--")
plt.legend()
plt.grid()
plt.show()
```
