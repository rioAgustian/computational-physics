```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

# kanvas
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# komponen statis
x = np.linspace(0, 10, 200)
y = np.linspace(0, 10, 200)
X, Y = np.meshgrid(x, y)

def f(x, y, t):
  a = np.cos(10*x - 2*t)
  b = np.sin(5*y - 5*t)
  c = a + b
  return c

# fungsi dari animasi dengan variabel frame
def wave(n):

  Z = f(X, Y, n)
  u = ax.plot_surface(X, Y, Z)
  ax.set_xlim(0, 10)
  ax.set_ylim(0, 10)
  ax.set_zlim(-4, 4)
  return u

# membuat animasi
graph = animation.FuncAnimation(fig, wave, frames=100, interval=20, blit=True)

# menampilkan  animasi yang telah dibuat
```
