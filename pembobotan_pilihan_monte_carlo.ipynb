Reaksi nuklir sifatnya probabilistik. Misalnya, terdapat tiga kemungkina reaksi: scattering (50%), capture (25%), dan fussion (25%). Kode ini memberikan kita cara bagaimana membuat pilihan dari 3 reaksi tersebut secara random, tetapi dengan pembobotan yang sudah kita tentukan.

```python
#cara pertama, menggunakan if-elif-else
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

n = 10000
r = np.random.default_rng().uniform(0,1,n)
reaction = []

for i in range(n):
  if r[i] >= 0 and r[i]<= 0.5:
    reaction.append("scattering")
  elif r[i] > 0.5 and r[i]<=0.75:
    reaction.append("capture")
  else:
    reaction.append("fussion")

print("Peluang scattering: ", reaction.count("scattering")/n)
print("Peluang capture: ", reaction.count("capture")/n)
print("Peluang fussion: ", reaction.count("fussion")/n)

pd.Series(reaction).value_counts(sort=False).plot(kind='bar')
```

```python
#cara kedua: menggunakan random.choices
reactions = ["scattering", "capture", "fussion"]
reaction = random.choices(reactions, weights=(0.5, 0.25, 0.25), k=10000)

print("Peluang scattering: ", reaction.count("scattering")/n)
print("Peluang capture: ", reaction.count("capture")/n)
print("Peluang fussion: ", reaction.count("fussion")/n)

pd.Series(reaction).value_counts(sort=False).plot(kind='bar')
```
